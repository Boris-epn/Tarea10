### a)

Para conocer si una matriz es no singular primero calcularemos su determinante
$$
determinante=\\
4*0*-3+(2*7*-2)+(6*3*-1)-(2*0*-2)-(-1*4*4)-(-3*3*2)\\
=0
$$
Por lo tanto la matriz es singular, y por ende no tiene una matriz inversa.

### b)

Se puede determinar la matriz inversa a través del método de gauss Jordan
$$
